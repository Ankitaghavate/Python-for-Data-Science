```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x=np.linspace(1,15,20)
```

```python
x
```

```python
y=x**2
y
```

```python
plt.ylabel("Square")
plt.xlabel("Number")
plt.plot(x,y,'r-')
plt.title("Number Square")
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'g')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])

fig=plt.figure()
axes=fig.add_axes([0.5,0.8,0.15,0.16])
axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Title")
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes1.plot(x,y)
axes1.set_title("Larger Plot")


axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig,axes=plt.subplots(nrows=1,ncols=6)
```

```python
fig2=plt.figure()
axes=fig2.add_axes([0.1,0.4,0.6,0.9])
axes.set_title('Title')
fig,axes=plt.subplots(nrows=1,ncols=5)
```

```python
fig1,axes1=plt.subplots(nrows=4,ncols=2)
```

```python
fig1,axes1=plt.subplots(nrows=1,ncols=2)
axes1[0].plot(x,y)
axes1[1].plot(y,x)


for current_axes in axes1:
    current_axes.plot(x,y)
```

## Figure Size and DPI

```python
fig=plt.figure(figsize=(4,3))
axes10=fig.add_axes([0,0.1,1.2,1.4])
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(7,4))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
fig.savefig('fig1.png',dpi=200)
```

tight_layout() helps adjust the subplot layout to avoid overlap.

```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(8,6))
ax[0].plot(x,y,label='X label')
ax[1].plot(y,x,'r',label='Y label')
fig.legend(loc=1)
fig.tight_layout()
fig.savefig('img1.jpg')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.3,0.5,0.7])
axes.plot(x,y,color='red',linewidth=2,alpha=0.5,linestyle='--',marker='o',markersize=10,markerfacecolor='blue',markeredgewidth=2,markeredgecolor='green')
#axes.plot(x,y,color='#3FF432')
#axes.plot(x,y,'orange')
```
