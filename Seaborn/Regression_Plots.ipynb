```python
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
```

sns.lmplot() creates a scatter plot with a linear regression line. It shows the relationship between two continuous variables and fits a straight line to the data. You can also add color to show different categories and create multiple plots using faceting.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue='sex',aspect=0.6)
```
