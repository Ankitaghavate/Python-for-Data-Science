Usage: Creates a pair plot.
    
Explanation: Displays scatter plots for every pair of numeric variables in a dataset, allowing for easy identification of relationships.


```python
sns.pairplot(tips)
```

Usage: Creates a regression plot.
    
Explanation: Plots data and fits a regression line, useful for visualizing relationships between variables and linear trends.

```python
sns.regplot(x="total_bill", y="tip", data=tips)
```

catplot

Usage: Creates categorical plots.
    
Explanation: A figure-level function for creating various types of categorical plots like bar plots, box plots, and violin plots.

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

displot

Usage: Creates a distribution plot.
    
Explanation: Visualizes the distribution of a variable using histograms or kernel density plots.


```python
sns.displot(tips['total_bill'])
```

sns.lmplot()

Usage: Creates a linear model plot.

Explanation: Fits and visualizes a linear regression model to the data.

```python
sns.lmplot(x="total_bill", y="tip", data=tips)
```

sns.jointplot()

Usage: Creates a joint plot.
    
Explanation: Displays the relationship between two variables along with their marginal distributions (histogram or KDE).

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

sns.countplot()

Usage: Creates a count plot.
    
Explanation: Displays the count of observations in each categorical bin using bars.

```python
sns.countplot(x='day',data=tips)
```

sns.kdeplot()

Usage: Creates a kernel density estimate plot.

Explanation: Visualizes the probability density function of a continuous variable, smoothed from the data points.

```python
sns.kdeplot(tips['total_bill'])
```

Usage: Creates a bar plot.
    
Explanation: Displays a summary of categorical data using rectangular bars, with the height representing the central tendency (mean) of a variable.

```python
sns.barplot(x='day',y='total_bill',data=tips)
```
