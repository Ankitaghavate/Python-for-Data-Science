we  convert 0 and 1 value into yes and No for Better Understand

```python
def conv(value):
  if value==1:
    return "Yes"
  else:
    return "No"

df["SeniorCitizen"]=df["SeniorCitizen"].apply(conv)
```

```python

# Compute counts and percentages
ct = pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize="index") * 100  # Convert to % of total

# Plot stacked bar chart
fig, ax = plt.subplots(figsize=(6, 4))
bottom = None

for churn_status in ct.columns:
    bars = ax.bar(ct.index, ct[churn_status], label=churn_status, bottom=bottom)
    bottom = ct[churn_status] if bottom is None else bottom + ct[churn_status]

    # Add text labels
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_y() + height/2,
                    f"{height:.1f}%", ha="center", va="center", fontsize=10, color="white")

# Labels and title
plt.xlabel("Senior Citizen (0 = No, 1 = Yes)")
plt.ylabel("Percentage")
plt.title("Churn by SeniorCitizen (Stacked %)")
plt.xticks([0, 1], ["No", "Yes"])
plt.legend(title="Churn")

plt.show()
```

```python
# Define the categorical columns
columns = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
           'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']

# Set the number of rows and columns for subplots
rows = (len(columns) + 2) // 3  # Adjusting for neat layout
fig, axes = plt.subplots(rows, 3, figsize=(15, rows * 4))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Create count plots for each categorical column
for i, col in enumerate(columns):
    sns.countplot(x=df[col], data=df, ax=axes[i], palette="Set2")
    axes[i].set_title(f"Count of {col}")
    axes[i].set_xlabel(col)
    axes[i].set_ylabel("Count")
    axes[i].set_xticklabels(axes[i].get_xticklabels(), rotation=45)  # Rotate for better visibility

# Adjust layout
plt.tight_layout()
plt.show()
```
